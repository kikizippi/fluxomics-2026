## 3. Comparison limitations between FBA and published 13C-MFA datasets

FBA and 13C-MFA are both used to study metabolic fluxes, but they do not produce the same type of output.

FBA predicts feasible flux distributions based on stoichiometric constraints and an objective function, while 13C-MFA estimates intracellular fluxes using isotope-labeling experiments and metabolic models.

For this reason, the comparison between the computational FBA results and the published 13C-MFA datasets is mainly qualitative and methodological, not a direct one-to-one numerical comparison.

## 4. Computational summary

This notebook reviewed selected published fluxomics and 13C-MFA datasets that are relevant to the FBA-based computational analysis of this project.

Four published datasets were selected because they are related to E. coli central carbon metabolism, carbon source effects, oxygen availability, or experimentally estimated intracellular fluxes.

The selected datasets were not used as direct numerical replacements for the FBA simulations. Instead, they were used as external published references for qualitative and methodological comparison.

The main FBA analysis in this project produces predicted flux behavior based on stoichiometric constraints, medium composition, and an objective function. In contrast, published 13C-MFA datasets provide experimentally constrained intracellular flux estimates based on isotope-labeling measurements.

For this reason, objective values and blocked reactions from FBA cannot be directly compared with 13C-MFA outputs. However, intracellular flux behavior, carbon source effects, and aerobic versus anaerobic metabolic differences can be compared qualitatively when the biological context is similar.

Overall, this review supports the structure of the project by connecting the computational FBA datasets with independent published fluxomics studies. It also clarifies the limitations of comparing predictive constraint-based models with experimental isotope-based flux estimation.
